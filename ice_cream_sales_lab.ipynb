# Question 2: sales v. temp

```python
data = pd.DataFrame({
   'Temperature (F)': [57.56, 61.52, 53.42, 59.36, 65.3, 71.78, 66.92, 77.18, 74.12, 64.58],
   'Sales': [215, 325, 185, 332, 406, 522, 412, 614, 544, 421]
})
sns.relplot(data=data, x='Temperature (F)', y='Sales', size='Sales')
```

### conclusion:
unsurprisingly, sales are higher when the temperature is higher.  The relationship is comparatively linear for the temperature range between 55* and 75*

## 3A: total units across all months

```python
data = pd.read_csv('Monthly Ice Cream Shop Sales by Flavor.csv')
fig, ax = plt.subplots()
ax.plot(data['Month'], data['Total Units'])
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.set_title('Sales by Month')
```

```python
fig, ax = plt.subplots()
for i, flavor in enumerate(data):
    if 0 < i < len(data.columns)-2:
        ax.plot(data['Month'], data[flavor], label=flavor)
ax.legend()
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.set_title('Sales by Month')
```
